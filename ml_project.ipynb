```python
df_drop=df.drop(columns=["id","date","floors","waterfront","view","condition","grade","sqft_basement","yr_built","yr_renovated","zipcode","lat","long"])
```

```python
df_drop.keys()
```

```python
df_drop.describe()
```

```python
df_corr=df_drop.corr()

np.ones_like(df_corr, dtype=np.bool)
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))
,
mask = mask[1:, :-1]
corr = df_corr.iloc[1:,:-1].copy()
sns.heatmap(corr,cmap="Blues",mask=mask, square=True,annot=True,
           linewidth=0.3, cbar_kws={"shrink": .8})
```
